```python
song_name = []
release_date = []
album = []
played_at =[]
id = []
artist =[]
for i in data['items']:
    song_name.append(i['track']['name'])
    release_date.append(i['track']['album']['release_date'])
    album.append(i['track']['album']['name'])
    artist.append(i['track']['album']['artists'][0]['name'])
    played_at.append(i['played_at'][0:10])
    id.append(i['played_at'])
    list_dict = {
        "song_name":song_name,
        "release_date":release_date,
        "album":album,
        "artist":artist,
        "played_at":played_at,
        "id":id,
    }
df = pd.DataFrame(list_dict)
```

```python
# create a database or path to the existing database
DATABASE_LOCATION = "sqlite:///tracks.sqlite"
engine = sqlalchemy.create_engine(DATABASE_LOCATION)
```

```python
#create a curson to take care of the SQL QUERY by connecting to the database created
conn = sqlite3.connect('tracks.sqlite')
cursor = conn.cursor()
#create a table
create_table = """CREATE TABLE SONGS (
                id VARCHAR(200),
                song_name VARCHAR(200),
                release_date VARCHAR(200),
                album VARCHAR(200),
                artist VARCHAR(200),
                played_at VARCHAR(200)
                )"""
cursor.execute(create_table)
```

```python
#Load data into the table 
try:
    df.to_sql('songs',engine,index = False,if_exists='append')
    print("data is loaded")
except:
    print("data is already loaded")
```
